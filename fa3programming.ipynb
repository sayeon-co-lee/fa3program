```python
#fa3prog2

# Accept the student's exam score
score = float(input("Enter the student's exam score: "))

# Check the condition and display the result
if score >= 75:
    print("Result: Passed")
else:
    print("Result: Failed")
```

```python
#fa3prog3

# Input student details
name = input("Enter the student's name: ")
score = float(input(f"Enter the score for {name}: "))

# Determine the letter grade based on metrics
if score >= 90 and score <= 100:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"

# Display the result
print("\n--- Grade Report ---")
print(f"Student: {name}")
print(f"Score:   {score}")
print(f"Grade:   {grade}")
```
